```python
unit_raw = df["unit"].astype(str).str.upper().str.strip()

# normalize multiple spaces
unit_raw = unit_raw.str.replace(r"\s+", " ", regex=True)

# extract leading number if it exists (e.g., 100 in "100 KG")
qty = unit_raw.str.extract(r"^(\d*\.?\d+)\s*", expand=False)

df["unit_qty"] = pd.to_numeric(qty, errors="coerce").fillna(1.0)

# remove the number part to get the unit text
df["unit_name"] = unit_raw.str.replace(r"^\d*\.?\d+\s*", "", regex=True).str.strip()

df[["unit","unit_qty","unit_name"]].head(10)
```

```python
df["unit_name"].value_counts().head(20)
```

```python
def to_kg(qty, unit_name):
    if unit_name in ["KG", "KGS"]:
        return qty
    if unit_name in ["G", "GRAM", "GRAMS"]:
        return qty / 1000
    return np.nan  # not a weight unit

df["qty_in_kg"] = df.apply(lambda r: to_kg(r["unit_qty"], r["unit_name"]), axis=1)

# price per kg (only where qty_in_kg is known)
df["price_per_kg"] = np.where(df["qty_in_kg"].notna(), df["price"] / df["qty_in_kg"], np.nan)

df[["unit","unit_qty","unit_name","qty_in_kg","price","price_per_kg"]].head(10)
```

```python
df["price_per_kg"].notna().mean()
```

```python
def to_liters(qty, unit_name):
    if unit_name in ["L", "LITRE", "LITER", "LITRES", "LITERS"]:
        return qty
    if unit_name in ["ML"]:
        return qty / 1000
    return np.nan

df["qty_in_l"] = df.apply(lambda r: to_liters(r["unit_qty"], r["unit_name"]), axis=1)
df["price_per_l"] = np.where(df["qty_in_l"].notna(), df["price"] / df["qty_in_l"], np.nan)
```

```python
def to_liters(qty, unit_name):
    if unit_name in ["L", "LITRE", "LITER", "LITRES", "LITERS"]:
        return qty
    if unit_name in ["ML"]:
        return qty / 1000
    return np.nan

df["qty_in_l"] = df.apply(lambda r: to_liters(r["unit_qty"], r["unit_name"]), axis=1)
df["price_per_l"] = np.where(df["qty_in_l"].notna(), df["price"] / df["qty_in_l"], np.nan)
```

```python
import numpy as np

# Clean unit_name
df["unit_name"] = df["unit_name"].astype(str).str.upper().str.strip()

# --------- WEIGHT: convert to KG ----------
df["qty_in_kg"] = np.nan

mask_kg = df["unit_name"].isin(["KG", "KGS"])
df.loc[mask_kg, "qty_in_kg"] = df.loc[mask_kg, "unit_qty"]

mask_g = df["unit_name"].isin(["G", "GRAM", "GRAMS"])
df.loc[mask_g, "qty_in_kg"] = df.loc[mask_g, "unit_qty"] / 1000

df["price_per_kg"] = np.where(df["qty_in_kg"].notna(), df["price"] / df["qty_in_kg"], np.nan)


# --------- VOLUME: convert to L ----------
df["qty_in_l"] = np.nan

mask_l = df["unit_name"].isin(["L", "LITRE", "LITER", "LITRES", "LITERS"])
df.loc[mask_l, "qty_in_l"] = df.loc[mask_l, "unit_qty"]

mask_ml = df["unit_name"].isin(["ML"])
df.loc[mask_ml, "qty_in_l"] = df.loc[mask_ml, "unit_qty"] / 1000

df["price_per_l"] = np.where(df["qty_in_l"].notna(), df["price"] / df["qty_in_l"], np.nan)

# Preview result
df[["unit", "unit_qty", "unit_name", "qty_in_kg", "price_per_kg", "qty_in_l", "price_per_l"]].head(20)
```

```python
df["unit_standard"] = df["unit_name"]  # start as original

# weight
df.loc[df["unit_name"].isin(["G","GRAM","GRAMS","KG","KGS"]), "unit_standard"] = "KG"

# volume
df.loc[df["unit_name"].isin(["ML","L","LITRE","LITER","LITRES","LITERS"]), "unit_standard"] = "L"

df[["unit_name","unit_standard"]].value_counts().head(20)
```

```python
df["qty_standard"] = np.nan

# for KG group
mask_weight = df["unit_standard"].eq("KG")
df.loc[mask_weight, "qty_standard"] = df.loc[mask_weight, "qty_in_kg"]

# for L group
mask_vol = df["unit_standard"].eq("L")
df.loc[mask_vol, "qty_standard"] = df.loc[mask_vol, "qty_in_l"]
```

```python
df["price_per_standard"] = np.where(
    df["qty_standard"].notna(),
    df["price"] / df["qty_standard"],
    np.nan
)
```

```python
df["unit_name"] = df["unit_name"].astype(str).str.upper().str.strip()

df["unit_group"] = "OTHER"

# WEIGHT group
df.loc[df["unit_name"].isin(["KG", "KGS", "G", "GRAM", "GRAMS"]), "unit_group"] = "WEIGHT"

# VOLUME group
df.loc[df["unit_name"].isin(["L", "LITRE", "LITER", "LITRES", "LITERS", "ML"]), "unit_group"] = "VOLUME"

# COUNT group (your UNIT, PCS, TUBERS idea)
df.loc[df["unit_name"].isin(["UNIT", "UNITS", "PCS", "PC", "PIECE", "PIECES", "TUBER", "TUBERS"]), "unit_group"] = "COUNT"

df["unit_group"].value_counts()
```

```python
df["price_per_count"] = np.where(
    df["unit_group"].eq("COUNT"),
    df["price"] / df["unit_qty"],
    np.nan
)

df[df["unit_group"]=="COUNT"][["unit","unit_qty","unit_name","price","price_per_count"]].head(20)
```

```python
df["analysis_price"] = np.nan

# For weight items -> use NGN per KG
df.loc[df["unit_group"].eq("WEIGHT"), "analysis_price"] = df["price_per_kg"]

# For volume items -> use NGN per L
df.loc[df["unit_group"].eq("VOLUME"), "analysis_price"] = df["price_per_l"]

# For count items -> use NGN per item
df.loc[df["unit_group"].eq("COUNT"), "analysis_price"] = df["price_per_count"]
```
